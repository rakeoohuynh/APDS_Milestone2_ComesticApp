```python
import pandas as pd
```

```python
df = pd.read_csv('processed.csv')
```

```python
group_df = df.groupby('product_title')['product_id']
```

```python
group_df = df.drop_duplicates(subset='product_id').reset_index(drop=True)
print(f"Unique products: {len(group_df)}")
group_df.head()
```

```python
# Select only these two columns
selected_columns = group_df[['product_id', 'product_title']]
print(selected_columns)
```

```python
# English comment: Sort the entire DataFrame by 'product_id'. 
# This keeps 'product_title' and all other columns aligned.
selected_columns = selected_columns.sort_values(by='product_id', ascending=True)
```

```python
selected_columns.to_csv('product.csv')
```
